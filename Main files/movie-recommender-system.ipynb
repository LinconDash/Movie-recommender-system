```python
# Columns to retrive are : genres,id,keywords,title,overview,cast,crew
movies = movies[['id','title','genres','keywords','overview','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
# Formatting the genres,keywords,cast,crew columns into right format
movies.iloc[0]['genres']
```

```python
# ast is a python module whih has a function called literal_eval that converts the string-list into a list
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# Fetch only the first 3 actors from the cast dictionary
def fetch3cast(obj):
    L = []
    count = 1
    for i in ast.literal_eval(obj):
        if count <= 3:
            L.append(i['name'])
            count += 1
        else:
            break
    return L       
```

```python
movies['cast'].apply(fetch3cast).iloc[0]
```

```python
movies['cast'] = movies['cast'].apply(fetch3cast)
```

```python
# For the crew column Lets fetch the director name only
def find_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(find_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
# Converting the overview to a list type
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
# Applying some transformations for removing space between words
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['overview'] = movies['overview'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# create a new column called Tags to concatenate genres,keywords,overview,cast,crew
movies['Tags'] = movies['overview'] +  movies['genres'] +  movies['keywords'] +  movies['cast'] +  movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['id','title','Tags']]
```

```python
new_df.head()
```

```python
# Since the Tags column is a list we have to create it as a string with space separated
new_df['Tags'] = new_df["Tags"].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df["Tags"][0]
```

```python
# Its recommended to change all the case into lower case
new_df['Tags'] = new_df['Tags'].apply(lambda x : x.lower())
new_df["Tags"][0]
```

```python
# Converting the Tags words to vectors (Vectorization) . Each movie will be vector .
# A movie is recommended when someone likes a movie , then the vectors close to that movie vector is recommended
# Also for words we will not consider stop words (and,for,to,also,etc...)
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words="english")
```

```python
# Lets convert all the movie tags to a matrix of shape(num_of_movies, 5000 words)
word_vectors = cv.fit_transform(new_df["Tags"]).toarray()
```

```python
# Lets have a look at the vector of avatar movie Tags
word_vectors[0]
```

```python
# We have a method to see all the 5000 words(that appears most frequently) that are used to create vectors
cv.get_feature_names()
```

```python
# Since we have same words that are quivalent to each other we will use a NLP technique called stemming
!pip install nltk
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
# Example of the stemming technique
stem("loving love loved")
```

```python
# so lets apply the stemming technique again in the Tags column of the new_df Dataframe to easily eliminate the repeated words.
print("Before Stemming : \n" + new_df["Tags"][0] + "\n")
print("After Stemming : \n" + stem(new_df["Tags"][0]))
```

```python
new_df["Tags"] = new_df['Tags'].apply(stem)
```

```python
# Now again do vectorization to eliminate repitition of the words
word_vectors = cv.fit_transform(new_df["Tags"]).toarray()
# word_vectors[0]
cv.get_feature_names()
```

```python
# So for the similarity of the vectors we have to find the cosine distance and not the euclidean distance.
# Because Euclidean distance fails with increase in dimensionality (Here it is 5000)
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity(word_vectors)
```

```python
# Each movie's similarity is calculated with every other movie
cosine_similarity(word_vectors).shape
```

```python
similarity_matrix = cosine_similarity(word_vectors)
```

```python
# Lets see the similarity of Avatar with everyother movies 
# Note : In the similarity matrix the diagonals will always be 1 since each movie is completely similar to itself
similarity_matrix[0]
```

```python
# Lets create a main function that takes a movie name and recommend 5 movies most similar to it
# In this function we will take the index of the movie and see into the similarity_matrix 
# and sort the similarity values and get the top 5 similar values and use it to recommend 5 movies
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity_matrix[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    # Lets print the movie names
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
# Since we have to sort them without using the index of the movies so we have to write a complex statement.
# Also using a lambda function to sort them accorting to similarity distances otherwise it will sort by index numbers
sorted(list(enumerate(similarity_matrix[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
# Testing: recommended movies for Avatar
recommend("Avatar")
```
